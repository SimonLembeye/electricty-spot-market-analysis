```python
def plot_graph(x, y, title="", xlabel="", ylabel=""):
    
    if len(x) != len(y):
        print("len(x) != len(y)")
        print(len(x), len(y))
        return
    
    x_up = []
    y_up = []
    x = np.array(x)
    y = np.array(y)
    
    for i in range(len(x)):
        if not np.isnan(x[i]) and not np.isnan(y[i]):
            x_up.append(x[i])
            y_up.append(y[i])
        
    
    x = np.array(x_up)
    y = np.array(y_up)
    
    m, b = np.polyfit(x, y, 1)

    fig = plt.figure()
    
    plt.scatter(x, y, s=0.1)
    plt.plot(x, b + m * x, 'r-')
    
    fig.suptitle(title, fontsize=18)
    plt.xlabel(xlabel, fontsize=14)
    plt.ylabel(ylabel, fontsize=14)
    
    plt.grid()
    plt.show()
    fig.savefig('temp.png', dpi=fig.dpi)

    print("slop:" + str(m))
    print("b:" + str(b))
```

```python
def get_values_and_production(country_df):

    values_and_production = {}

    type_of_energy = np.delete(country_df.columns.values, [0, 1])
    prices = country_df["price"].values

    for energy_name in type_of_energy:

        values_and_production[energy_name] = {
            "production": 0,
            "value": 0,
        }

        values = country_df[energy_name].values
        for j in range(len(prices)):
            if not np.isnan(values[j]):
                price = prices[j]
                prod = values[j]

                values_and_production[energy_name]["production"] += prod / 1000
                values_and_production[energy_name]["value"] += (prod * price / 1000)


    for energy_name in type_of_energy:
        values_and_production[energy_name]["value"] /= values_and_production[energy_name]["production"] 
        
        print(energy_name + " - " + "prod: " + str(values_and_production[energy_name]["production"]) + ", avg price: " + str(values_and_production[energy_name]["value"]))
```

```python
def plot_price_duration_curve(prices):

    x = [i / len(prices) for i in range(len(prices))]

    prices = np.array(prices)
    prices.sort()

    fig = plt.figure()

    plt.plot(x, prices[::-1], 'r-')
    fig.suptitle("Price/Duration Curve", fontsize=18)
    plt.grid()
    plt.show()
    fig.savefig('temp.png', dpi=fig.dpi)
```

```python
def negative_prices_analysis(country_df):
    
    number_of_hours = 0
    avg_negative_price = 0.
    
    prices = country_df["price"].values
    
    for i in range(len(prices)):
        
        price = prices[i]
        
        if price < 0:
            number_of_hours += 1
            avg_negative_price += price
    
    if number_of_hours > 0:
        avg_negative_price /= number_of_hours
    
    print("Number of hours with negative prices: " + str(number_of_hours))
    print("Average Negative Price: " + str(avg_negative_price))
```

```python
def model_price_function(vector, net_load):
    res = 0
    
    for i in range(len(vector)):
        res += (vector[len(vector) - i - 1] * pow(net_load, i))
        
    return res
```

```python
def get_country_generator(country_df, degree=10, plot_figure=True):
    x = country_df['load'].values - (country_df["solar"].values + country_df["wind_offshore"].values + country_df["wind_onshore"].values)
    y = country_df["price"].values

    if len(x) != len(y):
        print("len(x) != len(y)")
        print(len(x), len(y))

    x_up = []
    y_up = []
    x = np.array(x)
    y = np.array(y)

    for i in range(len(x)):
        if not np.isnan(x[i]) and not np.isnan(y[i]):
            x_up.append(x[i])
            y_up.append(y[i])


    x = np.array(x_up)
    y = np.array(y_up)

    vector = np.polyfit(x, y, degree)
    
    if plot_figure:

        fig = plt.figure()

        plt.scatter(x, y, s=0.1)

        plt.plot(x, model_price_function(vector, x), 'r-')
        
        fig.suptitle("France Model", fontsize=18)
        plt.xlabel("Net Load (GW)", fontsize=14)
        plt.ylabel("Price (â‚¬/MWh)", fontsize=14)

        plt.grid()
        plt.show()
        fig.savefig('temp.png', dpi=fig.dpi)
    
    return vector


get_country_generator(france, degree=5)
```

```python
def get_values_and_production_with_model(country_df, vector):
    
    print()
    print("Model predictions")
    
    energies = ["load", "nuclear", "wind_onshore", "wind_offshore", "solar", "gas"]

    values_and_production = {}

    type_of_energy = np.delete(country_df.columns.values, [0, 1])
    
    net_load = country_df['load'].values - (country_df["solar"].values + country_df["wind_offshore"].values + country_df["wind_onshore"].values)
    
    for energy_name in energies:

        values_and_production[energy_name] = {
            "production": 0,
            "value": 0,
        }

        values = country_df[energy_name].values
        for j in range(len(net_load)):
            if not np.isnan(values[j]) and not np.isnan(net_load[j]):
                price = model_price_function(vector, net_load[j])
                prod = values[j]
                

                values_and_production[energy_name]["production"] += prod / 1000
                values_and_production[energy_name]["value"] += (prod * price / 1000)


    for energy_name in energies:
        values_and_production[energy_name]["value"] /= values_and_production[energy_name]["production"] 
        
        print(energy_name + " - " + "prod: " + str(values_and_production[energy_name]["production"]) + ", avg price: " + str(values_and_production[energy_name]["value"]))
        
    return values_and_production
```

```python
vector = get_country_generator(germany, degree=5)
model_predictions = get_values_and_production_with_model(germany, vector)
```

```python
def get_price_variations(country_df, country_installed_capacites, extra_capacities):

    new_solar = country_df["solar_load_factor"].values * (country_installed_capacites["solar"] + extra_capacities["solar"])
    new_wind_onshore = country_df["wind_onshore_load_factor"].values * (country_installed_capacites["wind_onshore"] + extra_capacities["wind_onshore"])
    new_wind_offshore = country_df["wind_offshore_load_factor"].values * (country_installed_capacites["wind_offshore"] + extra_capacities["wind_offshore"])

    new_energy_values = {
        "load": country_df["load"].values,
        "solar": new_solar,
        "wind_onshore": new_wind_onshore,
        "wind_offshore": new_wind_offshore,
        "nuclear": country_df["nuclear"].values,
    }

    new_net_load = country_df['load'].values - (new_solar + new_wind_onshore + new_wind_offshore)

    values_and_production = {}

    energies = ["load", "nuclear", "wind_onshore", "wind_offshore", "solar"]
    
    vector = get_country_generator(country_df, degree=7)
    model_predictions = get_values_and_production_with_model(country_df, vector)
    
    for energy_name in energies:

        values_and_production[energy_name] = {
            "production": 0,
            "value": 0,
        }

        values = new_energy_values[energy_name]
        for j in range(len(new_net_load)):
            if not np.isnan(values[j]) and not np.isnan(new_net_load[j]):
                price = model_price_function(vector, new_net_load[j])
                prod = values[j]
                values_and_production[energy_name]["production"] += prod / 1000
                values_and_production[energy_name]["value"] += (prod * price / 1000)

    
    print()
    print("# Variations")
    for energy_name in energies:
        values_and_production[energy_name]["value"] /= values_and_production[energy_name]["production"] 
        
        prod_variation = values_and_production[energy_name]['production'] - model_predictions[energy_name]['production']
        price_variation = values_and_production[energy_name]['value'] - model_predictions[energy_name]['value']
        print(energy_name + " - " + "prod: " + str(values_and_production[energy_name]["production"]) + ", avg price: " + str(values_and_production[energy_name]["value"]))
        print("Production variation:", prod_variation)
        print("Price variation:", price_variation)
        print()
        
    prod_variation = values_and_production["load"]['production'] - model_predictions["load"]['production']
    price_variation = values_and_production["load"]['value'] - model_predictions["load"]['value']
    
    price_variation_solar = values_and_production["solar"]['value'] - model_predictions["solar"]['value']
    price_variation_wind_on = values_and_production["wind_onshore"]['value'] - model_predictions["wind_onshore"]['value']
    price_variation_wind_off = values_and_production["wind_offshore"]['value'] - model_predictions["wind_offshore"]['value']
    
    
    print("Gain:", abs(price_variation * model_predictions["load"]['production'] * 1000))
    print("Cost Solar:", price_variation_solar * model_predictions["solar"]['production'] * 1000)
    print("Cost Wind Onshore:", price_variation_wind_on * model_predictions["wind_onshore"]['production'] * 1000)
    print("Cost Wind Offshore:", price_variation_wind_off * model_predictions["wind_offshore"]['production'] * 1000)
    
    print("Cost Extra prod: ", - 1000 * (values_and_production["solar"]['production'] - model_predictions["solar"]['production']) * (65 - model_predictions["solar"]['value']))
```
